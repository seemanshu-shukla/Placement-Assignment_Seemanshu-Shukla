```python
# Importing pythonic libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#Reading the dataset stored in csv format
df = pd.read_csv("nasa_transformed_data.csv")

#Displaying the dataset as a pandas dataframe
df
```

```python
#Creating a new column to store Year value

df['Only_Year'] = np.nan
```

```python
#Flattening 'year' column by splitting data using '-' and storing the 0th index resultant in 'Only_Year' column

for i in range(1000):
    if isinstance(df['year'][i], str) == True:
        df['Only_Year'][i]=int((df['year'][i].split("-"))[0])
    else:
        continue
```

```python
df
```

```python
#Get all the Earth meteorites that fell before the year 2000

earth_meteorites_before_year_2000 = df[df["Only_Year"] < 2000]
earth_meteorites_before_year_2000[['name']]
```

From above it is clear that there are around 929 records representing Earth meteorites in the dataset that fell before year 2000

```python
#Visualizing the earth meteorites co-ordinates who fell before the year 2000

plt.hist(earth_meteorites_before_year_2000["Only_Year"], bins=35, edgecolor="black")
plt.title("Earth meteorites fell before 2000")
plt.xlabel("Year")
plt.ylabel("Count")
```

```python
#df['fall'].value_counts()
```

```python
#Get all the earth meteorites co-ordinates who fell before the year 1970

earth_meteorites_coordinates_fell_before_1970 = df[df["Only_Year"] < 1970]
earth_meteorites_coordinates_fell_before_1970[['geolocation.coordinates']]
```

From above it is clear that there are around 780 records representing earth meteorites coordinates in the dataset that fell before year 2000

```python
#Visualizing the earth meteorites co-ordinates who fell before the year 1970

'''
From observing the dataset it is clear that reclong represents the longitude which is equivalent to x co-ordinate whereas,
reclat represents the latitude which is equicalent to the y co-ordinate.

'''
plt.figure(figsize=(15,5))
plt.scatter(earth_meteorites_coordinates_fell_before_1970["reclong"], earth_meteorites_coordinates_fell_before_1970["reclat"],
            s=10)
plt.title("Earth meteorites coordinates that fell before 1970")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
