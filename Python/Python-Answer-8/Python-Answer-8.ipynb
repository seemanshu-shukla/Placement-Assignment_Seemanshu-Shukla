```python
# Importing pythonic libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#Reading the dataset stored in csv format
df = pd.read_csv("TV_Show_Tranformed_Data.csv")

#Displaying the dataset as a pandas dataframe
df
```

```python
df.drop(['Unnamed: 0'] , axis = 1, inplace = True)
df
```

```python
#Insight 1: Average Rating for each season

season_ratings = df.groupby('season')['rating.average'].mean()
season_ratings
```

```python
# Plot 1: Ratings for each season
plt.figure(figsize=(3,3))
season_ratings.plot(kind='bar', edgecolor='black')
plt.title("Season Ratings")
plt.xlabel("Seasons")
plt.ylabel("Average Rating")
plt.show()
```

```python
#Insight 2: Get all the episode names, whose average rating is more than 8 for every season

Rating_more_than_8 = df[df['rating.average'] > 8]

#Printing the episode names, whose average rating is more than 8 for every season
Rating_more_than_8[['name', 'season']]
```

```python
#Plot 2: Plotting highly rated (> 8) episode count for every season

df2 = df
df2['Is Rating > 8'] = df['rating.average'] > 8
sns.countplot(x='Is Rating > 8', hue='season',data=df2)
```

```python
# Convert the 'airdate' column to datetime type so that we can carry out Insight 3
df['airdate'] = pd.to_datetime(df['airdate'])
```

```python
#Insight 3: Get all the episode names that aired before May 2019

before_may_2019_episodes = df[df['airdate'] < pd.to_datetime('2019-05-01')]
before_may_2019_episodes[['name']]
```

```python
# Plot 3: Episode names that aired before May 2019
plt.figure(figsize=(15,10))
plt.scatter(x=before_may_2019_episodes['airdate'], y=before_may_2019_episodes['name'])
plt.title("Episodes Aired Before May 2019")
plt.xlabel("Airdate")
plt.ylabel("Episode Name")
plt.show()
```

```python
# Insight 4: Get the episode name from each season with the highest and lowest rating

#Note: idxmax() and idxmin() returns a series with the first index of the rows/columns with the highest values.

#high_rated_episodes = df.groupby('season').apply(lambda x: x.loc[x['rating.average'].idxmax()]['name'])
#low_rated_episodes = df.groupby('season').apply(lambda x: x.loc[x['rating.average'].idxmin()]['name'])

#Note: both idxmax()/idxmin() and max()/min() will be returning the first occurence. But there can be multiple occurences. So,
#to fetch multiple occurence we will neeed to write a separate function.

hre = []
hre_name = []
high_rated_episodes_per_season = df.groupby('season')['rating.average'].max()

def high_rate_episode_name(high_rated_episodes_per_season):
    for i in high_rated_episodes_per_season:
        hre.append(i)
    for i in range(len(hre)):
        for j in range(len(df)):
            if (df['season'][j] == i+1) and (df['rating.average'][j] == hre[i]):
                hre_name.append((df['season'][j],df['number'][j],df['name'][j],df['rating.average'][j]))
            else:
                continue
    return(hre_name)
        
high_rated = high_rate_episode_name(high_rated_episodes_per_season)

print("="*100)
print("Episode Name From Each Season With The Highest Rating:\n")
for i in high_rated:
    print(f"Season number = {i[0]}; Episode number = {i[1]}; Episode name = {i[2]}; Rating = {i[3]}")
print("="*100)
#In output for highest rated episode name we can clearly observe that there are 2 highly rated episodes in season 3 which
#would have otherwise get missed in case we just fetched name using .max() or .idxmax() methods.

lre = []
lre_name=[]
low_rated_episodes_per_season = df.groupby('season')['rating.average'].min()
def low_rate_episode_name(low_rated_episodes_per_season):
    for i in low_rated_episodes_per_season:
        lre.append(i)
    for i in range(len(lre)):
        for j in range(len(df)):
            if (df['season'][j] == i+1) and (df['rating.average'][j] == lre[i]):
                lre_name.append((df['season'][j],df['number'][j],df['name'][j],df['rating.average'][j]))
            else:
                continue
    return(lre_name)
        
low_rated = low_rate_episode_name(low_rated_episodes_per_season)

print("="*100)
print("Episode Name From Each Season With The Lowest Rating:\n")
for i in low_rated:
    print(f"Season number = {i[0]}; Episode number = {i[1]}; Episode name = {i[2]}; Rating = {i[3]}")
print("="*100)
#In output for lowest rated episode name we can clearly observe that there are 2 low rated episodes in season 3 which
#would have otherwise get missed in case we just fetched name using .max() or .idxmax() methods.
```

```python
# Plot 4: Get the episode name from each season with the highest and lowest rating

x_high=[]
y_high=[]
for i in high_rated:
    x_high.append(i[2])
    y_high.append(i[3])
    
x_low=[]
y_low=[]
for j in low_rated:
    x_low.append(j[2])
    y_low.append(j[3])

plt.figure(figsize=(20,50))
plt.subplot(6,2,1)
plt.bar(x_high,y_high)
plt.title("Highest Rated Episodes")
plt.xlabel("Episode Names")
plt.ylabel("Ratings")
plt.subplot(6,2,2)
plt.bar(x_low,y_low)
plt.title("Lowest Rated Episodes")
plt.xlabel("Episode Names")
plt.ylabel("Ratings")

plt.show()
```

```python
#Insight 5: Get the summary for the most popular ( ratings ) episode in every season

#In Insight 4 we have already fetched details of most popular episodes in every season which is stored in the variable with
#name high_rated

#Episode name can be used as an unique idenifier to fetch the respective summary

summary_popular=[]

for i in range(len(df)):
    for j in high_rated:
        if df['name'][i] == j[2]:
            summary_popular.append((j[0],j[1],j[2],df['summary'][i]))

print("Summary of most popular episodes in every season:\n")
for i in summary_popular:
    print("="*100)
    print(f"Season number = {i[0]}\nEpisode number = {i[1]}\nEpisode name = {i[2]}\nSummary = {i[3]}")
    print("="*100)
```

```python
# Plot 5: Get the summary for the most popular ( ratings ) episode in every season

x=[]
y=[]

#We will represent Summary on y axis instead of x axis since we will have more vertical space in y axis as compared to x axis 
for i in summary_popular:
    x.append(i[2])    
    y.append(i[3])


plt.figure()
plt.bar(x,y)
plt.title("Summary of Most Popular Episodes")
plt.xlabel("Episode Names")
plt.ylabel("Summary")
plt.xticks(rotation=90)
plt.show()
```
