#### a.  INTERPRETATION:
The range indicates the spread of data between the maximum and minimum values in each dataset. In this case, the range of 'Blood Pressure Before' is 28 mmHg, while the range of 'Blood Pressure After' is 23 mmHg.

The variance and standard deviation provide measures of the average squared deviation from the mean. In the 'Blood Pressure Before' dataset, the variance is approximately 43.101 mmHg², and the standard deviation is approximately 6.56 mmHg. Similarly, in the 'Blood Pressure After' dataset, the variance is approximately 46.9704 mmHg², and the standard deviation is approximately 6.85 mmHg.

A higher range, variance, and standard deviation indicate a greater dispersion or variability in the dataset. Therefore, 'Blood Pressure Before' shows slightly higher dispersion compared to 'Blood Pressure After' as indicated by the larger values of range, variance, and standard deviation.As variance keeps on increasing , spread will be also increasing.

Confidence Interval:
To calculate the confidence interval, we need to consider the sample size (n), standard deviation (s), and the desired confidence level. 

n = 100
Standard Error (SE) = σ / sqrt(n)
Margin of Error (ME) = t-value * SE
Confidence Interval = Mean ± ME


Since the standard deviation of the population is unknown, we'll use the t-distribution for calculating the confidence interval.

confidence level of 95% 

degrees of freedom (df) = n - 1 = 100-1 =  99

t-value = 1.984 (using t-distribution table)

#### Confidence Interval for 'Blood Pressure Before':

Standard Error (SE) = 6.56 / sqrt(100) = 0.656 mmHg

Margin of Error (ME) = 1.984 * 0.656 ≈ 1.301 mmHg

Confidence Interval for 'Blood Pressure Before' ≈ 133.91 ± 1.301 mmHg

Confidence Interval = (132.60075849172767, 135.21924150827232)

#### Confidence Interval for 'Blood Pressure After':

Standard Error (SE) ≈ 0.685 mmHg

Margin of Error (ME) ≈ 1.359 mmHg

Confidence Interval for 'Blood Pressure After' ≈ 128.36 ± 1.359 mmHg

Confidence Interval = (126.99326697818071, 129.7267330218193)

#### d. Correlation Coefficient and Significance:


Correlation Coefficient:
Using the formula for the Pearson correlation coefficient:

r = cov(x,y)/ (σx* σy)

Where x and y represent the values in the 'Blood Pressure Before' and 'Blood Pressure After' columns, and  σ represents the standard deviation.

correlation_coefficient = 1



```python
correlation_coefficient, p_value = stats.pearsonr(Blood_Pressure_Before, Blood_Pressure_Before)
```

```python
correlation_coefficient
```

```python
p_value
```

```python
alpha = 0.01
if p_value < alpha:
    print("The correlation is significant at the 1% level.")
else:
    print("The correlation is not significant at the 1% level.")
```
