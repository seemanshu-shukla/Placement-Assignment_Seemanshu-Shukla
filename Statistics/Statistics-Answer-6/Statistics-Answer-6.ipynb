```python
#Importing required libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#Reading dataset
df = pd.read_csv('data.csv')
df
```

```python
#Checking columns name

df.columns
```

```python
BP_Before=df[' Blood Pressure Before (mmHg)']
```

```python
BP_After=df[' Blood Pressure After (mmHg)']
```

```python
#Calculating the difference

diff=BP_Before-BP_After
```

```python
#Ploting histogram and visualizing the normality
plt.figure(figsize=(3,4))
sns.histplot(diff, kde=True)
plt.xlabel('Change in Blood Pressure')
plt.ylabel('Frequency')
plt.title('Histogram of Blood Pressure Changes')
plt.show()
```

```python
stats.probplot(diff, dist="norm", plot=plt)
plt.title('Q-Q Plot')
plt.show()
```

#### Observation:
 - By observing histogram we can say that the data is normally distributed since there is no/negligible skewness in the data distribution.

 - Q-Q plot is slightly or negatively deviating from the straight line thus making it a perfect fit for Normal Distribution.

Therefore, the change in blood pressure follows a normal distribution.
